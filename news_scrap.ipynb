```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

existing_csv_file = 'output.csv'

dfinal = pd.DataFrame(columns=['heading', 'description'])
df_existing = pd.read_csv(existing_csv_file)
print(df_existing.shape)
```

```python
final = pd.DataFrame(columns=['heading', 'description'])
list=['india','entertainment','education','sports','politics','lifestyle','opinion']
for k in list:
  for j in range(1, 100):
      webpage = requests.get('https://indianexpress.com/section/{}/page/{}/'.format(k,j)).text
      soup = BeautifulSoup(webpage, 'lxml')
      articles = soup.find_all('div', class_='articles')

      for i in articles:
          try:
              description = i.find('p').text.strip()
          except:
              description = None

          try:
              heading = i.find('h2').text.strip()
          except:
              heading = None


          final = final.append({'heading': heading, 'description': description}, ignore_index=True)
```

```python
final.shape
combined_df = pd.concat([final,df_existing], ignore_index=True)
unique_df = combined_df.drop_duplicates()
print(unique_df.shape)
unique_df.to_csv('output.csv', index=False)
```
